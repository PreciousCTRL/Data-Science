## How to clean a json file

```python
reviews_df = pd.DataFrame.from_records(review_df['reviews'])
reviews_df.head()
```

# Access a SAS data set

```python
eval_df = pd.read_sas('../data/techsupportevals.sas7bdat')

eval_df.info()

customers_df = pd.read_sas('../snowflake/customers.sas7bdat')
```

```python
eval_df
```

## Reading and Accessing Parquet files


```python
gcs_key = '../keys'

customer_churn_df = pd.read_parquet('../gcs/customer_churn_data.parquet')
```

# JOINING DATA

```python
df = pd.merge(customer_churn_df, customers_df, on = "custId", how = 'inner').drop(columns = 'custId')
```

```python
df = df.merge(subs_df, on = "customerSubscrCode", how = 'inner').drop(columns = 'customerSubscrCode')
```

```python
df = df.merge(eval_df, on = "ID", how = 'inner')
```

```python
df = df.merge(reviews_df, on = "reviewId", how = 'left').drop(columns = 'reviewId')
```

```python
df.columns
```

```python
df.head()
```

```python
df['DemHomeOwnerCode'] = df['DemHomeOwnerCode'].str.decode('utf-8')
df['customerGender'] = df['customerGender'].str.decode('utf-8')
```

```python
df['demHomeOwner'] = df['DemHomeOwnerCode'].map({"U": 'Unknown', "H":'HomeOwner'})

df.drop(columns="DemHomeOwnerCode", inplace = True)
```

```python
df.head()
```

```python
df['demHomeOwner'].value_counts()
```

```python
df['customerAge'] = ((pd.Timestamp.now() - pd.to_datetime(df['birthDate'])).dt.days/365.25)
```

```python
df['customerAge'] =df['customerAge'].apply(lambda x: int(x) if pd.notnull(x) else np.nan)
```

```python
df.drop(columns='birthDate', inplace = True)
```

```python
df['AvgPurchasePerAd'] = df['AvgPurchaseAmount12']/df['intAdExposureCount12']
```

```python
df.to_csv('../data/customer_churn.csv')
```
